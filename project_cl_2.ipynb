```python
black_friday_revenue = orderline_orders[(orderline_orders["date"].dt.month == 11) & (orderline_orders["date"].dt.weekday == 4)]["total_paid"].sum()
christmas_revenue = orderline_orders[(orderline_orders["date"].dt.month == 12) & (orderline_orders["date"].dt.day.between(20, 26))]["total_paid"].sum()
```

```python
# Identify Black Friday (4th Friday of November)
black_friday_revenue = orderline_orders[
    (orderline_orders["date"].dt.month == 11) &
    (orderline_orders["date"].dt.weekday == 4) &
    (orderline_orders["date"].dt.day >= 23) & (orderline_orders["date"].dt.day <= 29)
]

# Identify Christmas revenue (20th to 26th December)
christmas_revenue = orderline_orders[
    (orderline_orders["date"].dt.month == 12) &
    (orderline_orders["date"].dt.day.between(20, 26))
]

# Create a boolean column
orderline_orders['month_1'] = orderline_orders['date'].isin(black_friday_revenue['date']) | \
                               orderline_orders['date'].isin(christmas_revenue['date'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grouping and summing the data
sum_by_month_false_dis = orderline_orders.loc[(orderline_orders['month_1'] == False) & (orderline_orders['unit_price_total'])] \
    .groupby('quarter', as_index=False)['unit_price_total'].sum().sort_values(by='unit_price_total', ascending=False)

sum_by_month_true_dis = orderline_orders.loc[(orderline_orders['month_1'] == True) & (orderline_orders['unit_price_total'])] \
    .groupby('quarter', as_index=False)['unit_price_total'].sum().sort_values(by='unit_price_total', ascending=False)

# Add a label to differentiate the data
sum_by_month_true_dis['label'] = 'Black Friday & Christmas'
sum_by_month_false_dis['label'] = 'Other Days'

# Combine both DataFrames
combined_df = pd.concat([sum_by_month_true_dis, sum_by_month_false_dis])

# Combine the data for Quarter 4
quarter_4_true = sum_by_month_true_dis[sum_by_month_true_dis['quarter'] == 4]['unit_price_total'].values[0]
quarter_4_false = sum_by_month_false_dis[sum_by_month_false_dis['quarter'] == 4]['unit_price_total'].values[0]

# Create a new DataFrame for Quarter 4 combined
quarter_4_combined = pd.DataFrame({
    'quarter': [4],
    'unit_price_total': [quarter_4_true + quarter_4_false],
    'label': ['Black Friday & Christmas + Other Days']
})

# Remove original Quarter 4 data from both DataFrames
sum_by_month_true_dis = sum_by_month_true_dis[sum_by_month_true_dis['quarter'] != 4]
sum_by_month_false_dis = sum_by_month_false_dis[sum_by_month_false_dis['quarter'] != 4]

# Combine the adjusted DataFrames and the new Quarter 4 combined data
combined_df = pd.concat([sum_by_month_true_dis, sum_by_month_false_dis, quarter_4_combined])

# Sort the combined DataFrame by 'unit_price_total'
combined_df = combined_df.sort_values(by='unit_price_total', ascending=False)

# Plot using Seaborn
plt.figure(figsize=(10, 10))
custom_colors = ['#4c78A8', '#F58518']
sns.barplot(x='quarter', y='unit_price_total', hue='label', data=combined_df, palette=custom_colors)
plt.title('Revenue Comparison: Black Friday & Christmas vs Other Days by Quarter')
plt.ylabel('Total Revenue (Millions)')
#plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Extract year and quarter for proper distinction
orderline_orders['year'] = orderline_orders['date'].dt.year
orderline_orders['quarter_label'] = 'Q' + orderline_orders['quarter'].astype(str) + ' ' + orderline_orders['year'].astype(str)

# Group and sum the data for month_1 = False
sum_by_month_false_dis = orderline_orders.loc[(orderline_orders['month_1'] == False) & (orderline_orders['unit_price_total'])]
summary_false = sum_by_month_false_dis.groupby('quarter_label', as_index=False)['unit_price_total'].sum()
summary_false['label'] = 'Other Days'

# Group and sum the data for month_1 = True
sum_by_month_true_dis = orderline_orders.loc[(orderline_orders['month_1'] == True) & (orderline_orders['unit_price_total'])]
summary_true = sum_by_month_true_dis.groupby('quarter_label', as_index=False)['unit_price_total'].sum()
summary_true['label'] = 'Black Friday & Christmas'

# Combine the data
combined_df = pd.concat([summary_true, summary_false])

# Plot using Seaborn
plt.figure(figsize=(12, 8))
custom_colors = ['#4c78A8', '#F58518']
sns.barplot(x='quarter_label', y='unit_price_total', hue='label', data=combined_df, palette=custom_colors)

plt.title('Revenue Comparison: Black Friday & Christmas vs Other Days by Quarter and Year')
plt.ylabel('Total Revenue (Millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
